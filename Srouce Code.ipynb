##Finding mean of each gene

```python
for i in range(df.shape[0]):
  df.loc[i,"cancer mean"]=np.mean(list(data1.iloc[i,1:27]))
  df.loc[i,"non-cancer mean"]=np.mean(list(data1.iloc[i,27:65]))
```

```python
df.head()
```

##Finding p value

```python
#null hypothesis - There is no difference between canerous and non cancerous gene

#What is meant by a P value?
#A p-value, or probability value, is a number
#describing how likely it is that your data would have occurred under the null hypothesis of your statistical test.

from scipy.stats import ttest_ind

# Loop over each row in the gene expression matrix
for i in range(data1.shape[0]):
    t_stat, p_val = ttest_ind(list(data1.iloc[i,1:27]), list(data1.iloc[i,27:65]),equal_var=False)
    df.loc[i,"p-value"]=p_val
    #df.loc[i,"t_stat"]=t_stat
```

```python
df.head(20)
```

```python

def func(arr,th):
  count=0
  for i in arr:
    if i<th:
      count+=1
  print("There are "+str(count)+" out of "+str(data1.shape[0])+" genes below p-value of "+str(th))
  return "................................................................"
```

```python
print(func(df['p-value'],0.05))
print(func(df['p-value'],0.06))
print(func(df['p-value'],0.07))
print(func(df['p-value'],0.08))
print(func(df['p-value'],0.09))
print(func(df['p-value'],0.1))
```

# Finding Adjusted-P value

###Bonferroni


```python
#Bonferroni

for i in range(df.shape[0]):
  df.loc[i,'bon'] = df.loc[i,'p-value'] * len(df)
```

```python
df.head()
```

###Benjamini and Hochberg

```python
df.sort_values('p-value',inplace=True)
df.reset_index(inplace=True,drop=True)
```

```python
df.head()
```

```python
#Benjamini and Hochberg adj p val
df['BH'] = df['p-value']*len(df)/(df.index+1)
```

```python
df.head(20)
```

```python
def func(arr,th):
  count=0
  for i in arr:
    if i<th:
      count+=1
  print("There are "+str(count)+" out of "+str(data1.shape[0])+" genes below adj p-value of "+str(th))
  return "."
```

```python
print(func(df['BH'],0.05))
print(func(df['BH'],0.06))
print(func(df['BH'],0.07))
print(func(df['BH'],0.08))
print(func(df['BH'],0.09))
print(func(df['BH'],0.1))
```
